```python
import pandas as pd

# Define file path
file_path = "/content/drive/My Drive/Colab Notebooks/ML Project/players_stats_by_season_full_details.csv"

# Load the dataset
df = pd.read_csv(file_path)

# Create NBA_Indication: 1 if League equals 'NBA', otherwise 0
df['NBA_Indication'] = df['League'].apply(lambda x: 1 if x == 'NBA' else 0)

# Create Regular_Season_Indication: 1 if Stage equals 'Regular_Season', otherwise 0
df['Regular_Season_Indication'] = df['Stage'].apply(lambda x: 1 if x == 'Regular_Season' else 0)

# Create International_Indication: 1 if Stage equals 'International', otherwise 0
df['International_Indication'] = df['Stage'].apply(lambda x: 1 if x == 'International' else 0)

# Create Playoffs_Indication: 1 if Stage equals 'Playoffs', otherwise 0
df['Playoffs_Indication'] = df['Stage'].apply(lambda x: 1 if x == 'Playoffs' else 0)

df.head()
```

```python
# Identify players with inconsistent birth_year data:
# (i.e. some rows have a non-null value and some rows are null)
inconsistent_birth_year = df.groupby('Player')['birth_year'].apply(lambda x: x.notnull().any() and x.isnull().any())

# Print the list of players with inconsistent birth_year values
players_inconsistent = inconsistent_birth_year[inconsistent_birth_year].index.tolist()
print("Players with inconsistent birth_year values:", players_inconsistent)
```

```python
def get_inconsistent_values(series):
    # If a player's birth_year column has at least one non-null and at least one null, return all unique values.
    if series.notnull().any() and series.isnull().any():
        return list(series.unique())
    else:
        return None

# Group by 'Player' and apply the function
inconsistent_birth_year = df.groupby('Player')['birth_year'].apply(get_inconsistent_values)

# Filter out the players without inconsistencies
inconsistent_birth_year = inconsistent_birth_year[inconsistent_birth_year.notnull()]

print("Players with inconsistent birth_year values and their values:")
print(inconsistent_birth_year)
```

```python
def get_inconsistent_values(series):
    # Drop null values and get unique values.
    unique_vals = series.dropna().unique()
    # If there are more than one unique non-null values, return them.
    if len(unique_vals) > 1:
        return list(unique_vals)
    else:
        return None

# Group by 'Player' and apply the function to the 'birth_year' column.
inconsistent_birth_year = df.groupby('Player')['birth_year'].apply(get_inconsistent_values)

# Filter out players that do not have inconsistencies.
inconsistent_birth_year = inconsistent_birth_year[inconsistent_birth_year.notnull()]

print("Players with inconsistent non-null birth_year values:")
print(inconsistent_birth_year)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def get_inconsistent_values(series):
    unique_vals = series.dropna().unique()
    if len(unique_vals) > 1:
        return list(unique_vals)
    return None

# Group by 'Player' and find inconsistencies
inconsistent_birth_year = df.groupby('Player')['birth_year'].apply(get_inconsistent_values)
inconsistent_birth_year = inconsistent_birth_year[inconsistent_birth_year.notnull()]

# Function to search Wikipedia for birth year
def get_birth_year_from_wikipedia(player_name):
    search_url = f"https://en.wikipedia.org/w/index.php?search={player_name} basketball"
    headers = {"User-Agent": "Mozilla/5.0"}  # To avoid being blocked

    response = requests.get(search_url, headers=headers)
    if response.status_code != 200:
        return None  # Failed to retrieve

    soup = BeautifulSoup(response.text, "html.parser")
    first_link = soup.find("a", href=True, text=player_name)  # Find exact match

    if first_link:
        page_url = f"https://en.wikipedia.org{first_link['href']}"
        response = requests.get(page_url, headers=headers)
        if response.status_code == 200:
            page_soup = BeautifulSoup(response.text, "html.parser")
            birth_year = extract_birth_year(page_soup)
            return birth_year
    return None  # No relevant page found

def extract_birth_year(soup):
    birth_info = soup.find("span", class_="bday")
    if birth_info:
        return int(birth_info.text[:4])  # Extract year only
    return None

# Add new columns
df["final_birth_year"] = df["birth_year"]
df["birth_year_source"] = "original"

# Update inconsistent players
for player in inconsistent_birth_year.index:
    wiki_birth_year = get_birth_year_from_wikipedia(player)

    if wiki_birth_year:
        df.loc[df["Player"] == player, "final_birth_year"] = wiki_birth_year
        df.loc[df["Player"] == player, "birth_year_source"] = "wikipedia"
    else:
        df.loc[df["Player"] == player, "final_birth_year"] = None
        df.loc[df["Player"] == player, "birth_year_source"] = "unknown"

print("Process completed! Updated birth years in final_birth_year column.")

```

Handling Inconsistent Height and Weight Data
We identified players with inconsistent height and weight values, selected the most common value for each, and marked whether the value was original or adjusted

```python
from collections import Counter

def get_most_common_value(series):
    counter = Counter(series.dropna())
    if counter:
        return counter.most_common(1)[0][0]  # The most frequently occurring value
    return None

# Update height and weight with the most common value for each player
df["final_height_cm"] = df.groupby("Player")["height_cm"].transform(get_most_common_value)
df["final_weight_kg"] = df.groupby("Player")["weight_kg"].transform(get_most_common_value)

# Source column: "original" if consistent, "fixed" if adjusted
df["height_source"] = df.apply(lambda row: "original" if row["height_cm"] == row["final_height_cm"] else "fixed", axis=1)
df["weight_source"] = df.apply(lambda row: "original" if row["weight_kg"] == row["final_weight_kg"] else "fixed", axis=1)

print("✅ Inconsistencies resolved! Values have been replaced with the most common ones.")

```

```python
import pickle

# Define aggregation rules
aggregation_rules = {
    "GP": "sum",
    "MIN": "sum",
    "FGM": "sum",
    "FGA": "sum",
    "3PM": "sum",
    "3PA": "sum",
    "FTM": "sum",
    "FTA": "sum",
    "TOV": "sum",
    "PF": "sum",
    "ORB": "sum",
    "DRB": "sum",
    "REB": "sum",
    "AST": "sum",
    "STL": "sum",
    "BLK": "sum",
    "PTS": "sum",
    "NBA_Indication": "max",
    "Regular_Season_Indication": "max",
    "International_Indication": "max",
    "Playoffs_Indication": "max",
    "final_birth_year": "first",
    "birth_year_source": "first",
    "final_height_cm": "first",
    "final_weight_kg": "first",
    "height_source": "first",
    "weight_source": "first"
}

# Perform aggregation
df_aggregated = df.groupby("Player").agg(aggregation_rules).reset_index()

# Convert categorical columns to dummy variables
df_aggregated = pd.get_dummies(df_aggregated, columns=["height_source", "birth_year_source", "weight_source"], dtype=int)

# Define output file paths
output_csv = "/content/drive/My Drive/Colab Notebooks/ML Project/NBA_cleaned.csv"
output_pkl = "/content/drive/My Drive/Colab Notebooks/ML Project/NBA_cleaned.pkl"

# Save the processed data
df_aggregated.to_csv(output_csv, index=False)
df_aggregated.to_pickle(output_pkl)

# Print confirmation and final shape
print(f"✅ Data saved successfully: {output_csv} and {output_pkl}")
print("📊 Final dataset shape:", df_aggregated.shape)
```
